# Seaborn - Excercises

Working with the famous **titanic data set** for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#### Creating the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.

```python
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
sns.distplot(titanic['fare'], kde=False, bins=30, color='red')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x="class", y="age", data=titanic, palette='Set2')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(titanic, col='sex')
g = g.map(sns.distplot, 'age', )
```

```python
g = sns.FacetGrid(titanic, col='sex')
g = g.map(plt.hist, 'age', )
```
