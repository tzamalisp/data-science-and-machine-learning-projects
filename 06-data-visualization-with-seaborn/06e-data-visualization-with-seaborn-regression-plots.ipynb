# Seaborn - Regression Plots

Seaborn has many built-in capabilities for regression plots, however regression will not be really discussed until the relevant machine learning section later on, so I will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot() >> means: linear model plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

### hue parameter: add a seperation by a categorical feature (here, 'sex')

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

### You can use also `maplotlib` features to customize the plot. E.g., the dots could me other kind of markers:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o', 'v']) # these markers come from matplotlib marker samples
```

### change the color of the palette

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Working with Markers

**lmplot** `kwargs` gets passed through to **regplot**, which is a more general form of `lmplot()`. `regplot` has a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in that dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the basic matplotlib arguments, in this case, *s* for size of a scatter plot. In general, you probably won't remember this off the top of your head, but instead you have to reference the documentation.

#### Seaborn passes matplotlib's libraries to lmplot (like 'scatter_kws' argument). Note that these features are passed as a dictionary.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid

You can add more variable separation through columns and rows with the use of a grid. Just indicate this with the 'col' or 'row' arguments instead of 'hue' that is used above:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted with the **size** and **aspect** parameters:

#### Aspect is the ratio between the height and the width

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
